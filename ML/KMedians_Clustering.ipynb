# K-Meadians Clustering in Python

#### Implementation using Numpy

```python
import numpy as np

def k_median(X, k, max_iterations=100):
    # Step 1: Initialization
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    
    for _ in range(max_iterations):
        # Step 2: Assignment
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        
        # Step 3: Update Medians
        for i in range(k):
            centroids[i] = np.median(X[labels == i], axis=0)
    
    return centroids, labels
```

#### Example usage

```python
# Assuming you have a dataset X
X = np.random.rand(100, 2)  # Replace this with your dataset

# Set the number of clusters (k)
k = 3

# Run the k-median algorithm
centroids, labels = k_median(X, k)

# Print the results
print("Centroids:")
print(centroids)
print("Cluster Labels:")
print(labels) 
```

#### Visualize clusters

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

for i in range(k):
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}')

plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', label='Centroids')
plt.title('K-Median Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
